#### Label Encoding the Object type Columns

workclass ,marital_status ,occupation need to be handled as num type

```python
# Labeling the catagories with integers
for col in train_set.columns:
    if train_set[col].dtypes == object: # if the column has categorical values
        l_unique = train_set[col].unique() # find the unique values
        v_unique = np.arange(len(l_unique)) # create a list of number from zero to the length of the I_unique values
        train_set[col].replace(to_replace=l_unique, value=v_unique, inplace=True) # replace the categorical values with numerical values
        train_set[col] = train_set[col].astype('int64') # change the type from int64 to int32
        
        """# same has been done for test data as well
        test[col].replace(to_replace=l_unique, value=v_unique, inplace=True)
        test[col] = test[col].astype('int')
        """
```

```python
X=train_set.drop('wage_class',axis=1)
```

```python
y= train_set['wage_class']
y
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from xgboost import XGBClassifier
```

```python
# fit model no training data
model = XGBClassifier(objective='binary:logistic')
model.fit(train_x, train_y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# cheking training accuracy
y_pred = model.predict(train_x)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(train_y,predictions)
accuracy
```

```python
# Now to increae the accuracy of the model, we'll do hyperparameter tuning using grid search
from sklearn.model_selection import GridSearchCV
```

```python
model.get_params
```

```python
param_grid={
   
    'learning_rate':[1,0.5,0.1,0.01,0.001],
    'max_depth': [3,5,10,20],
    'n_estimators':[10,50,100,200]
}
```

```python
grid= GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid, verbose=3)
```

```python
grid.fit(train_x,train_y)
```

```python
# To  find the parameters givingmaximum accuracy
grid.best_params_
```

```python
# Create new model using the same parameters
new_model=XGBClassifier(learning_rate= 0.1, max_depth= 20, n_estimators= 200)
new_model.fit(train_x, train_y)
```

```python
y_pred_new = new_model.predict(test_x)
predictions_new = [round(value) for value in y_pred_new]
accuracy_new = accuracy_score(test_y,predictions_new)
accuracy_new
```
